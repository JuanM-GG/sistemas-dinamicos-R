# Desarrollo de modelo matemático 

En este libro analizaremos un biorreactor continuo en el cual se encuentran tres especies: sustrato, biomasa y producto. Se supone que en el flujo de alimentación al biorreactor solo se encuentra presente sustrato.

## Balances de materia

### Balance de materia total 

$$acumulación\ de\ materia\ =\ entrada\ por\ flujo\ -\ salida\ por\ flujo$$

$$\frac{d(\rho V)}{dt} = F_{in} \rho - F_{out} \rho$$

Considerando que la densidad, $\rho$, es constante y que en un reactor continuo el $F_{int} = F_{out}$

$$\frac{dV}{dt} = 0$$

En un reactor continuo el volumen permanece constante 

### Balance de materia para la biomasa

Ya que supisimos que no hay biomasa en el flujo de alimentación se tiene 

$$acumulación\ de\ biomasa\ =\ generación\ -\ salida\ por\ flujo$$


$$\frac{d(xV)}{dt} = r_x V - F x$$

Considerando que el volumen es constante y que la tasa de dilución se define como 

$$D = \frac{F}{V}$$

la cual es el inverso del tiempo de residencia y tiene unidades de $timepo^{-1}$

Se tiene que 

$$\frac{dx}{dt} = rx - D x$$

### Balance de materia para el sustrato 

$$acumulación\ de\ materia\ =\ entrada\ por\ flujo -\ salida\ por\ flujo -\ consumo$$ 

$$\frac{d(sV)}{dt} = F s_f - F s - r_s V$$ 
$$\frac{ds}{dt} = D (s_f - s) - r_s $$

### Balance de materia para el producto 

Al igual que la biomasa, no se encuntra producto presente en el flujo de alimentación

$$acumulación\ de\ producto\ =\ generación\ -\ salida\ por\ flujo$$

$$\frac{d(pV)}{dt} = r_p V - F p$$

$$\frac{dp}{dt} = r_p  - D p$$


## Tasa de crecimiento 

La velocidad de crecimiento de biomasa, $r_x$, se expresa comunmente como 

$$r_x = \mu(s) x$$

Donde $\mu(s)$ suele ser una función del sustrato. Existen diversas expresiones para representar esta función. Algunas de ellas consideran efectos de inhibición por sustrato o por producto. En el caso de inhibición por producto la función depende además del producto, es decir, $\mu(s,p)$. Para nuestro análisis consideramos la siguiente expresión que considera dicho efecto de inhibición por parte del producto 

$$\mu(s,p) =  \mu_{max} \left( \frac{s}{k_s + s}  \right) \left( \frac{k_{p} }{k_p + p}  \right) $$

Donde $\mu_{max}$ es la tasa máxima de crecimiento, $k_s$ represente la concentración de sustrato a la cual se alcanza la mitad de la tasa máxima de crecimiento, y $k_p$ es la concetración de producto a la cual la tasa máxima de crecimiento de reduce a la mitad. 

## Rendimientos 

Para relacionar la velocidad de crecimiento de biomasa, $r_x$, con las velocidades de consumo de sustrato, $r_s$, y generación de producto $r_p$, se emplean los rendimientos biomasa-sustrato y producto-biomasa, respectivamente. Estos rendimientos se definen como 

Rendimiento biomasa-sustrato 

$$Y_{xs} = \frac{r_x}{r_s}$$

Rendimiento producto-biomasa

$$Y_{px} = \frac{r_p}{r_x}$$

Donde $Y_{xs}$ y $Y_{px}$ son constantes. 

Considerando lo anterior, reescribimos las ecuaciones (1)-(3) para obtener el sistema de ecuaciones diferenciales que describen el proceso 

$$
\begin{aligned}

\frac{dx}{dt} &= r_x - D x \\
\frac{ds}{dt} &= D (s_f - s) - \frac{1}{Y_{xs}} r_x \\
\frac{dp}{dt} &= Y_{px} r_x - D p 

\end{aligned}
$$

siendo 

$$r_x = \mu_{max} \left(\frac{s}{k_s + s} \right) \left( \frac{k_p}{k_p + p} \right) x $$

# Análisis de estabilidad

Para determinar los puntos críticos de resulve el siguiente sistema de ecuaciones 

$$\frac{dx}{dt} = 0 \hspace{1cm} \rightarrow \hspace{1cm} D = \mu(s,p)$$


$$\frac{ds}{dt} = 0 \hspace{1cm} \rightarrow \hspace{1cm} x = Y_{px}(s_f - s)$$


$$\frac{dp}{dt} = 0 \hspace{1cm} \rightarrow \hspace{1cm} p = Y_{px} x$$

De esta manera se ecuentran tres puntos de equilibrio 

$$P_1: \hspace{1cm} \bar{s} = s_f, \hspace{1cm} \bar{x} = 0, \hspace{1cm} \bar{p} = 0$$

Los otros dos se encuentrar al resolver el polinomio de segundo grado 

$$s^2 + \left( \frac{\mu_{max} k_p}{Y_{px} Y_{xs} D} + k_s - \frac{k_p}{Y_{px} Y_{xs}} - s_f \right)s - \frac{k_s k_p}{Y_{px} Y_{xs}} - k_s s_f = 0$$

Al obtener la soluciones $\bar{s}$ sustituimos en (1) y (2) para encontrar los respectivos valores $\bar{x}$ y $\bar{p}$
